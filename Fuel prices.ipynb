```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv("wfp_food_prices_nigeria.csv", low_memory = False)
```

```python
data
```

```python
import warnings
warnings.filterwarnings(action='ignore')
data_fuel = data[1:][data["cmname"] == "Fuel (petrol-gasoline) - Retail"]
```

```python
data_fuel
```

```python
data_fuel.drop(columns = ["date", "cmname", "unit", "category", "currency", "country", "admname", "adm1id", "mktname"\
                         , "cmid", "ptid", "umid", "catid", "sn", "default"], inplace = True)
```

```python
data_fuel['mktid'] = pd.to_numeric(data_fuel['mktid'], downcast='integer')
```

```python
data_fuel
```

```python
data_fuel['mktid'] = pd.to_datetime(data_fuel['mktid'], format='%Y')
```

```python
data_fuel
```

```python
indexedData_fuel = data_fuel.set_index(data_fuel["mktid"])
indexedData_fuel
```

```python
indexedData_fuel.drop(columns="mktid", inplace=True)
```

```python
indexedData_fuel["price"] =pd.to_numeric(indexedData_fuel['price'], downcast='integer')
```

```python
indexedData_fuel.head(30)
```

```python
plt.plot(indexedData_fuel)
plt.savefig("plot.png")
plt.show()
```

```python
from statsmodels.tsa.stattools import adfuller

print("Results of Dickey Fuller Test")
dftest = adfuller(indexedData_fuel["price"], autolag = "AIC")
dfoutput = pd.Series(dftest[0:4], index= ["Test Statistics", "p-value", "#Lags Used", "Number of Observations Used"])

for key, value in dftest[4].items():
    dfoutput["Critical values (%s)"%key] = value
    
print(dfoutput)
```
